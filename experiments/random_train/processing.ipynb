```python
import json
import numpy as np
from matplotlib import pyplot as plt
```

```python
train_size = [10, 20, 50, 100, 500, 1000, 2000, 5000]

n = len(train_size)

eucl_tr_hits1 = np.zeros((n, 20))
eucl_tr_hits10 = np.zeros((n, 20))
eucl_ts_hits1 = np.zeros((n, 20))
eucl_ts_hits10 = np.zeros((n, 20))

hyp_tr_hits1 = np.zeros((n, 20))
hyp_tr_hits10 = np.zeros((n, 20))
hyp_ts_hits1 = np.zeros((n, 20))
hyp_ts_hits10 = np.zeros((n, 20))


for i, size in enumerate(train_size):
    for j in range(20):
        with open(f'results/{size}/{j+1}_hyperbolic_res.json') as fp:
            eucl_res = json.load(fp)
        with open(f'results/{size}/{j+1}_euclidean_res.json') as fp:
            hyp_res = json.load(fp)
        eucl_tr_hits1[i][j] = eucl_res['tr_hits1']
        eucl_tr_hits10[i][j] = eucl_res['tr_hits10']
        eucl_ts_hits1[i][j] = eucl_res['tr_hits1']
        eucl_ts_hits10[i][j] = eucl_res['tr_hits10']
        
        hyp_tr_hits1[i][j] = hyp_res['tr_hits1']
        hyp_tr_hits10[i][j] = hyp_res['tr_hits10']
        hyp_ts_hits1[i][j] = hyp_res['tr_hits1']
        hyp_ts_hits10[i][j] = hyp_res['tr_hits10']
        

eucl_tr_hits1_mean = np.mean(eucl_tr_hits1, axis=1)
eucl_tr_hits10_mean = np.mean(eucl_tr_hits10, axis=1)
eucl_ts_hits1_mean = np.mean(eucl_ts_hits1, axis=1)
eucl_ts_hits10_mean = np.mean(eucl_ts_hits10, axis=1)

hyp_tr_hits1_mean = np.mean(hyp_tr_hits1, axis=1)
hyp_tr_hits10_mean = np.mean(hyp_tr_hits10, axis=1)
hyp_ts_hits1_mean = np.mean(hyp_ts_hits1, axis=1)
hyp_ts_hits10_mean = np.mean(hyp_ts_hits10, axis=1)

eucl_tr_hits1_std = np.std(eucl_tr_hits1, axis=1)
eucl_tr_hits10_std = np.std(eucl_tr_hits10, axis=1)
eucl_ts_hits1_std = np.std(eucl_ts_hits1, axis=1)
eucl_ts_hits10_std = np.std(eucl_ts_hits10, axis=1)

hyp_tr_hits1_std = np.std(hyp_tr_hits1, axis=1)
hyp_tr_hits10_std = np.std(hyp_tr_hits10, axis=1)
hyp_ts_hits1_std = np.std(hyp_ts_hits1, axis=1)
hyp_ts_hits10_std = np.std(hyp_ts_hits10, axis=1)

```

```python
plt.figure(figsize=(14, 14))
plt.subplot(2, 2, 1)
plt.errorbar(train_size, eucl_tr_hits1_mean, eucl_tr_hits1_std, label='eucl_tr')
plt.errorbar(train_size, hyp_tr_hits1_mean, hyp_tr_hits1_std, label='hyp_tr')
plt.title('tr @ 1')
plt.xlabel('train_size')
plt.legend()

plt.subplot(2, 2, 2)
plt.errorbar(train_size, eucl_ts_hits1_mean, eucl_ts_hits1_std, label='eucl_ts')
plt.errorbar(train_size, hyp_ts_hits1_mean, hyp_ts_hits1_std, label='hyp_ts')
plt.title('ts @ 1')
plt.xlabel('train_size')
plt.legend()

plt.subplot(2, 2, 3)
plt.errorbar(train_size, eucl_tr_hits10_mean, eucl_tr_hits10_std, label='eucl_tr')
plt.errorbar(train_size, hyp_tr_hits10_mean, hyp_tr_hits10_std, label='hyp_tr')
plt.title('tr @ 10')
plt.xlabel('train_size')
plt.legend()

plt.subplot(2, 2, 4)
plt.errorbar(train_size, eucl_ts_hits10_mean, eucl_ts_hits10_std, label='eucl_ts')
plt.errorbar(train_size, hyp_ts_hits10_mean, hyp_ts_hits10_std, label='hyp_ts')
plt.title('ts @ 10')
plt.xlabel('train_size')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(14, 14))
plt.subplot(2, 2, 1)
plt.errorbar(train_size, eucl_tr_hits1_mean, eucl_tr_hits1_std, label='@ 1')
plt.errorbar(train_size, eucl_tr_hits10_mean, eucl_tr_hits10_std, label='@ 10')
plt.title('eucl train')
plt.xlabel('train_size')
plt.legend()

plt.subplot(2, 2, 2)
plt.errorbar(train_size, eucl_ts_hits1_mean, eucl_ts_hits1_std, label='@ 1')
plt.errorbar(train_size, eucl_ts_hits10_mean, eucl_ts_hits10_std, label='@ 10')
plt.title('eucl test')
plt.xlabel('train_size')
plt.legend()

plt.subplot(2, 2, 3)
plt.errorbar(train_size, hyp_tr_hits1_mean, hyp_tr_hits1_std, label='@ 1')
plt.errorbar(train_size, hyp_tr_hits10_mean, hyp_tr_hits10_std, label='@ 10')
plt.title('hyp train')
plt.xlabel('hyp train')
plt.legend()

plt.subplot(2, 2, 4)
plt.errorbar(train_size, hyp_ts_hits1_mean, hyp_ts_hits1_std, label='@ 1')
plt.errorbar(train_size, hyp_ts_hits10_mean, hyp_ts_hits10_std, label='@ 10')
plt.title('hyp test')
plt.xlabel('train_size')
plt.legend()

plt.show()
```
